```python
import numpy as np
```

```python

with open('crafting_times.csv','r') as infile:
    header = infile.readline().rstrip().split(',')
    data = []
    for line in infile:
        data.append([float(s) for s in line.rstrip().split(',')])
    data = np.array(data)
```

```python
Y = data[:,0]
X = np.hstack([np.ones([data.shape[0],1]),data[:,1:]])
for c in range(data[:,1:].shape[1]):
    if np.max(data[:,1+c]) != 1:
        X = np.hstack([X,data[:,1+c:1+c+1]*data[:,1+c:1+c+1]])
        print(np.max(data[:,1+c:1+c+1]*data[:,1+c:1+c+1]))
print(X.shape)
B = np.linalg.lstsq(X,Y)[0]
#B = np.linalg.inv(np.dot(X.T,X))
#B =  np.dot(B, X.T)
#B =  np.dot(B , Y)
```

```python
Yhat = np.dot(X,B)
residual  = Y-Yhat
```

```python
import matplotlib.pyplot as plt

plt.plot(Y,residual,'x')
plt.show()

plt.plot(Y,Yhat,'x')
plt.show()

print(np.mean(np.sqrt(residual*residual)))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation



model = Sequential([
    Dense(32, input_shape=(X.shape[1],)),
    Activation('relu'),
    Dense(32),
    Activation('relu'),
    Dense(32),
    Activation('relu'),
    Dense(1),
])

model = Sequential([
    Dense(32, input_shape=(X.shape[1],)),
    Activation('relu'),
    Dense(1),
])
```

```python

model.compile(optimizer='rmsprop',
              loss='mse')
```

```python
model.fit(X, Y, epochs=200, batch_size=4)
```

```python
Yhat = model.predict(X)
```

```python

residual  = Y[:]-Yhat[:,0]

plt.plot(Y,residual,'x')
plt.show()

plt.plot(Y,Yhat,'x')
plt.show()

print(np.mean(residual*residual))
```
